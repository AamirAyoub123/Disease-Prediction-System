```python
from sklearn.linear_model import LogisticRegression
```

**Importing The Dependencies**

```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

Loading the breast cancer  dataset into a Pandas

**Initialize a LabelEncoder instance.**

```python
label_encoder = LabelEncoder()
```

Transform the 'diagnosis' column into numerical labels (e.g., convert 'M' and 'B' into 1 and 0, respectively)

```python
labels=label_encoder.fit_transform(cancer_dataset.diagnosis)
```

Add a new column 'target' to the DataFrame, containing the numerical labels for the 'diagnosis' column.

```python
cancer_dataset['diagnosis']=labels
```

Transform the 'diagnosis' column into numerical labels (replacing the original values with 1 and 0)

```python
cancer_dataset.head()
```

Printing the count of unique values in the 'diagnosis' column of the breast cancer dataset

*  **0 = M = Malignant**
*  **1 = B = Benign**

```python
cancer_dataset['diagnosis'].value_counts()
```

**Separating the features (data) from the target labels (diagnosis)**

```python
features_data = cancer_dataset.drop(columns=['diagnosis','Unnamed: 32'],axis=1)
diagnosis_data = cancer_dataset['diagnosis']
```

```python
print(features_data)
```

```python
print(diagnosis_data)
```

Standardizing the data by scaling the features to have zero mean and unit variance.

```python
scaler = StandardScaler()
scaler.fit(features_data)
```

```python
standardized_data = scaler.transform(features_data)
print(standardized_data)
```

```python
features_data = standardized_data
diagnosis_data = cancer_dataset['diagnosis']
```

```python
print(features_data)
print(diagnosis_data)
```

Splitting the data into training and testing sets with an 80-20 split.

```python
features_data_train, features_data_test, diagnosis_data_train,diagnosis_data_test = train_test_split(features_data, diagnosis_data, test_size=0.2,
                                                                                             random_state=2)
```

```python
print(features_data.shape,features_data_train.shape, features_data_test.shape)
```

**Training The Model**

Initialize a Logistic Regression Model.

```python
model = LogisticRegression()
```

Training the Logistic Regression using the training data

```python
model.fit(features_data_train,diagnosis_data_train)
```

**Model Evaluation** : *Accuracy Score*

Accuracy Score On The Training Data.

```python
features_train_prediction = model.predict(features_data_train)
training_data_accuracy = accuracy_score(features_train_prediction,diagnosis_data_train)
print("Accuracy on the training data : ",training_data_accuracy)
```

Accuracy Score On The Test Data

```python
features_test_prediction = model.predict(features_data_test)
test_data_prediction = accuracy_score(features_test_prediction,diagnosis_data_test)
print("Accuracy on the test data" , test_data_prediction)
```

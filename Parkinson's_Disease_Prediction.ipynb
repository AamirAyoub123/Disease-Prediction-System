**Importing The Dependencies**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

Loading the Parkinson's dataset into a Pandas

```python
parkinsons_data=pd.read_csv('/content/parkinsons_data.csv')
```

Grouping the data based on the target variable(status).

```python
parkinsons_data.groupby('status').mean(numeric_only=True)
```

**Separating the features (data) from the target labels (status).**

```python
features_data=parkinsons_data.drop(columns=['name','status'],axis=1)
target_data=parkinsons_data['status']
```

```python
print(features_data)
```

```python
print(target_data)
```

Splitting the data into training and testing sets with an 80-20 split.

```python
features_train,features_test,target_train,target_test=train_test_split(features_data,target_data,
                                                                       test_size=0.2,random_state=2)
```

```python
print(features_data.shape,features_train.shape,features_test.shape)
```

Standardizing the data by scaling the features to have zero mean and unit variance.

```python
scaler = StandardScaler()
```

```python
scaler.fit(features_train)
```

```python
features_train = scaler.transform(features_train)

features_test = scaler.transform(features_test)
```

```python
print(features_train)
```

**Training The Model**

Initializing the Support Vector Machine (SVM) classifier with a linear kernel.

```python
classifier = svm.SVC(kernel='linear')
```

Training the Support Vector Machine classifier using the training data.

```python
classifier.fit(features_train, target_train)
```

**Model Evaluation** :
*Accuracy Score*

Accuracy Score On The Training Data

```python
features_train_prediction = classifier.predict(features_train)
training_data_accuracy=accuracy_score(features_train_prediction,target_train)
print(f"Accuracy score of the training data : {training_data_accuracy} ")
```

Accuracy Score On The Test Data

```python
features_test_prediction = classifier.predict(features_test)
test_data_accuracy=accuracy_score(features_test_prediction,target_test)
print(f"Accuracy score of the test data : {test_data_accuracy}")
```
